```python
# Check the collections if existing
collections = qdrant_client.get_collections().collections
existing_names = {c.name for c in collections}
existing_names
```

```python
# If collections do not exist then create else skip
if COLLECTION_NAME not in existing_names:
    print(f"Creating collection: {COLLECTION_NAME}")
    qdrant_client.create_collection(
        collection_name=COLLECTION_NAME,
        vectors_config=models.VectorParams(size=VECTOR_SIZE, distance=models.Distance.COSINE),
    )
else:
    print(f"Collection '{COLLECTION_NAME}' already exists (reusing).")
```

```python
# import libraries
import os
import numpy as np
from PIL import Image
from uuid import uuid4
```

```python
# Function to index image data inside the VDB
def index_image (image_path, category=None):
    img_embed = embedder.embed_image([image_path])[0]
    emb=np.array(img_embed).tolist()

    payload = {
        "filename": os.path.basename(image_path),
        "path":image_path,
        "category": category
    }

    qdrant_client.upsert(
        collection_name=COLLECTION_NAME,
        points=[
            models.PointStruct(
                id=str(uuid4()),
                vector=emb,
                payload=payload
            )
        ]
    )
    print(f"Indexed image: {image_path}")
    # print(f"Indexed â†’ {image_path}")
```

```python
# Image path f a croc
croc_img_path = str(images_root/"animal"/"crocodile.jpeg")
# croc_img_path
```

```python
# function to index files in the folder
def index_folder (root_folder):
    exts = (".jpg", ".jpeg", ".png", ".webp")
    for dirpath, _, files in os.walk(root_folder):
        category = os.path.basename(dirpath)
        for f in files:
            if f.lower().endswith(exts):
                image_path = os.path.join(dirpath, f)
                index_image(image_path, category=category)
```

#### Now lets perform the retrival operation

#### Text --> Image retrieval

```python
# Text to image search function
def search_text(query, k=5):
    results = qdrant_client.query_points (
        collection_name=COLLECTION_NAME,
        query=query,
        limit=k,
        with_payload=True
    )
    return results
```

```python
# Query 
query = "an image of a crocodile"
```

```python
# Pass the embedded query to our function above search text
response = search_text(embedder.embed_query(query))
response
```

```python
# Fetching the results
for point in response.points:
    print(point.payload, "score=", point.score)
    # print(f"Score: {r.score:.3f} â†’ {r.payload['path']}")
```

#### Image --> Image Retrieval

```python
# Image to image retrieval
def search_by_image(image_path, k=5):
    img_embed = embedder.embed_image([image_path])[0]
    results = qdrant_client.query_points(
        collection_name=COLLECTION_NAME,
        query=img_embed,
        limit=k,
        with_payload=True
    )
    return results
```

```python
# Image path
query_image = croc_img_path
```

```python
# Pass the image path to the function
result = search_by_image(query_image)
```

```python
# Fetch the result
for point in result.points:
    print(point.payload, "score=", point.score)
```
